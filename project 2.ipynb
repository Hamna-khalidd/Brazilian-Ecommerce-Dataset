# Top 10 best-selling product categories

```python
query="""with product_sales as (select order_items.product_id,payments.payment_value from order_items join payments
on order_items.order_id=payments.order_id)
select product_dataset.product_category_name,round(sum(product_sales.payment_value),2) as sales
from product_dataset join product_sales
on product_dataset.product_id=product_sales.product_id
group by product_dataset.product_category_name
order by sales desc"""

cur.execute(query)
data = cur.fetchall()
data
df=pd.DataFrame(data,columns=["Category","Sales"])
df.head(10)
```

# 	Which states generate the most revenue?

```python
query="""
with a as(select customer_dataset.customer_state,order_dataset.order_id  from customer_dataset join order_dataset 
ON customer_dataset.customer_id = order_dataset.customer_id)
select a.customer_state,round(sum(payments.payment_value),2) as sales from a join payments
on a.order_id=payments.order_id
group by a.customer_state
order by sales desc"""
cur.execute(query)
data = cur.fetchall()
data
df=pd.DataFrame(data,columns=["state","sales"])
df
            
```

# Average delivery time per product category

```python
querry="""select product_dataset.product_category_name,avg(day(order_dataset.order_estimated_delivery_date)) as avg_delivery_time
from product_dataset join order_items on product_dataset.product_id=order_items.product_id
join order_dataset on order_dataset.order_id=order_items.order_id
group by product_dataset.product_category_name
order by avg_delivery_time desc"""
cur.execute(query)
data=cur.fetchall()
data
df=pd.DataFrame(data,columns=["month","total_orders"])
df
```

# Monthly order volume trend

```python
query="""select date_format(order_purchase_timestamp, "%Y-%m")as order_month,count(order_id) as total_orders
from order_dataset where order_purchase_timestamp is not null
group by order_month
order by order_month"""
cur.execute(query)
data=cur.fetchall()
data
df=pd.DataFrame(data,columns=["month","total_orders"])
df
```
